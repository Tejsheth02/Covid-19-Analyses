```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
corona_dataset = pd.read_csv("C:\Covid-19 proj\covid19_Confirmed_dataset.csv")
corona_dataset.head()
```

```python
corona_dataset.shape
```

```python
corona_dataset.drop(["Lat","Long","Province/State"] ,axis = 1, inplace=True)
corona_dataset.head()
```

```python
corona_dataset_aggregated = corona_dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()

```

```python
corona_dataset_aggregated.shape
```

```python
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Afghanistan"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["China"].plot()
```

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates=[]
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
```

```python
max_infection_rates
```

```python
corona_dataset_aggregated["max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rates"])
```

```python
corona_data.head()
```

```python
deaths = pd.read_csv("C:\Covid-19 proj\covid19_deaths_dataset.csv")
```

```python
deaths.head()
```

```python
columns = ["Province/State","Lat","Long",]
```

```python
deaths.drop(columns , axis = 1, inplace = True)
```

```python
deaths.head()
```

```python
deaths.set_index('Country/Region', inplace = True)
```

```python
deaths.head()
```

```python
corona_data.shape
```

```python
deaths.shape
```

```python
deaths_aggregated = deaths.groupby("Country/Region").sum()
```

```python
deaths_aggregated.head()
```

```python
countries = list(deaths_aggregated.index)
total_deaths=[]
for c in countries :
    total_deaths.append(deaths_aggregated.loc[c].diff().max())
```

```python
total_deaths
```

```python
deaths_aggregated["Total_deaths"] = total_deaths
```

```python
deaths_data = pd.DataFrame(deaths_aggregated["Total_deaths"])
```

```python
deaths_data.head()
```

```python
data2 = corona_data.join(deaths_data,how="inner")
```

```python
data2.head()
```
